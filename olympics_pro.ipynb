```python
import numpy as np
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dt = pd.read_csv('C:/Users/PC/Documents/Olympics_Project/data/athlete_events.csv')
region = pd.read_csv('C:/Users/PC/Documents/Olympics_Project/data/noc_regions.csv')
```

```python
dt.head()
```

```python
dt.isnull().sum()
```

```python
#replace mean into NaN values
dt['Age'] = dt['Age'].fillna(dt['Age'].mean())
dt['Weight'] = dt['Weight'].fillna(dt['Weight'].mean())
dt['Height'] = dt['Height'].fillna(dt['Height'].mean())
```

```python
dt.isnull().sum()
```

```python
#Separate Summer and Winter tables
sm = dt[dt['Season'] == 'Summer']
sm = sm.drop(['ID'], axis='columns')
```

```python
wt = dt[dt['Season'] == 'Winter']
wt = wt.drop(['ID'], axis='columns')
```

```python
#Summer Athletes Participation 

fig, ax = plt.subplots(1, figsize=(24, 6))
ax.plot(sm[['Name', 'Year']].
        drop_duplicates(
        subset = ['Name', 'Year']).
        groupby('Year').
        count(),
        marker='o',
        color = 'tab:red')

ax.set_ylabel('Number of Athletes', fontsize=16)
ax.set_xlim([1895, 2017])
ax.set_xticks(np.arange(1896, 2017, 4))
ax.set_yticks(np.arange(0, 14000, 2000))
ax.set_ylim([-500, 12500])

ax.axvspan(1914, 1918, color='grey', alpha=0.15)
ax.axvspan(1939, 1945, color='grey', alpha=0.15)
ax.annotate("WWI", xy=(1915, 10000), xycoords='data', fontsize='12')
ax.annotate('WWII', xy=(1940, 10000), xycoords='data', fontsize='14')
ax.annotate('1906 Intercalated', xy=(1906, 1500), xycoords='data',
            xytext=(1906, 3500), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            )
ax.annotate("1952 Helsinki\nThe first Thailand's\n Participation", xy=(1952, 4800), xycoords='data',
            xytext=(1952, 1840), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            )
ax.annotate("1976 Montreal\nThe first Thailand's Medal\n(Bronze)", xy=(1976, 5600), xycoords='data',
            xytext=(1976, 3920), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            )
ax.annotate("1996 Atlanta\nThe first Thailand's Gold Medal", xy=(1996, 9800), xycoords='data',
            xytext=(1996, 8410), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            )
            

plt.title('Number of Athletes Participation at the Summer Olympics', fontsize=32)
plt.show()
```

```python
#Winter Athletes Participation 

fig, ax2 = plt.subplots(1, figsize=(24, 6))
ax2.plot(wt[['Name', 'Year']].
        drop_duplicates(
        subset = ['Name', 'Year']).
        groupby('Year').
        count(),
        marker='o',
        color = 'tab:blue')

ax2.set_ylabel('Number of Athletes', fontsize=16)
ax2.set_xlim([1923, 2015])
ax2.set_xticks(np.arange(1924, 2018, 4))
ax2.set_yticks(np.arange(0, 3200, 300))
ax.set_ylim([-500, 5000])

ax2.axvspan(1914, 1918, color='grey', alpha=0.15)
ax2.axvspan(1939, 1945, color='grey', alpha=0.15)
ax2.annotate('WWII', xy=(1940, 2400), xycoords='data', fontsize='20')

plt.title('Number of Athletes Participation at the Winter Olympics', fontsize=32)
plt.show()
```
