```python
import numpy as np
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt = pd.read_csv('C:/Users/PC/Documents/Olympics_Project/data/athlete_events.csv')
region = pd.read_csv('C:/Users/PC/Documents/Olympics_Project/data/noc_regions.csv')
```

```python
dt.head()
```

```python
dt.isnull().sum()
```

```python
#replace mean into NaN values
dt['Age'] = dt['Age'].fillna(dt['Age'].mean())
dt['Weight'] = dt['Weight'].fillna(dt['Weight'].mean())
dt['Height'] = dt['Height'].fillna(dt['Height'].mean())
```

```python
dt.isnull().sum()
```

```python
#Separate Summer and Winter tables
sm = dt[dt['Season'] == 'Summer']
sm = sm.drop(['ID'], axis='columns')
wt = dt[dt['Season'] == 'Winter']
wt = wt.drop(['ID'], axis='columns')
```

```python
#Summer Athletes Participation 

fig, ax = plt.subplots(1, figsize=(24, 6))
ax.plot(sm[['Name', 'Year']].
        drop_duplicates(
        subset = ['Name', 'Year']).
        groupby('Year').
        count(),
        marker='o',
        color = 'tab:red')

ax.set_ylabel('Number of Athletes', fontsize=16)
ax.set_xlim([1895, 2017])
ax.set_xticks(np.arange(1896, 2017, 4))
ax.set_yticks(np.arange(0, 14000, 2000))
ax.set_ylim([-500, 12500])

ax.axvspan(1914, 1918, color='grey', alpha=0.15)
ax.axvspan(1939, 1945, color='grey', alpha=0.15)
ax.annotate("WWI", xy=(1915, 10000), xycoords='data', fontsize='14')
ax.annotate('WWII', xy=(1940, 10000), xycoords='data', fontsize='14')
ax.annotate('1906 Intercalated', xy=(1906, 1500), xycoords='data',
            xytext=(1906, 3500), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            )
ax.annotate("1952 Helsinki\nThe first Thailand's\n Participation", xy=(1952, 4800), xycoords='data',
            xytext=(1952, 1840), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            )
ax.annotate("1976 Montreal\nThe first Thailand's Medal\n(Bronze)", xy=(1976, 5600), xycoords='data',
            xytext=(1976, 3920), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            )
ax.annotate("1996 Atlanta\nThe first Thailand's Gold Medal", xy=(1996, 9800), xycoords='data',
            xytext=(1996, 8410), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            )

plt.title('Number of Athletes Participation at the Summer Olympics', fontsize=24)
plt.savefig('summer_par.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Winter Athletes Participation 

fig, ax2 = plt.subplots(1, figsize=(24, 6))
ax2.plot(wt[['Name', 'Year']].
        drop_duplicates(
        subset = ['Name', 'Year']).
        groupby('Year').
        count(),
        marker='o',
        color = 'tab:blue')

ax2.set_ylabel('Number of Athletes', fontsize=16)
ax2.set_xlim([1923, 2015])
ax2.set_xticks(np.arange(1924, 2018, 4))
ax2.set_yticks(np.arange(0, 3200, 300))
ax.set_ylim([-500, 5000])

ax2.axvspan(1914, 1918, color='grey', alpha=0.15)
ax2.axvspan(1939, 1945, color='grey', alpha=0.15)
ax2.annotate('WWII', xy=(1940, 2400), xycoords='data', fontsize='20')

plt.title('Number of Athlete Participation at the Winter Olympics', fontsize=24)
plt.savefig('winter_par.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Merged Continent
dt2 = pd.merge(dt, region, on='NOC', how='left')
dt2.head()
```

```python
menInOlympics = dt2[(dt2.Sex == 'M') & (dt.Season == 'Summer')]
womenInOlympics = dt2[(dt2.Sex == 'F') & (dt.Season == 'Summer')]
```

```python
#Summer Participation by Sex
sns.set(style='darkgrid')
plt.figure(figsize=(20, 10))
sns.countplot(x='Year',hue='Sex', data=sm, palette=['C0', 'tab:pink'])
plt.title('Athlete Participation at the Summer Olympics by Sex', fontsize=24)
plt.ylabel('Number of Athletes', fontsize=16)
plt.savefig('athlete_par_bysex.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Number of Sport
fig, ax3= plt.subplots(1, figsize=(24, 6))
ax3.plot(sm[['Sport','Year']].
         drop_duplicates(
             subset=['Sport','Year']).
         groupby('Year').
         count(),
         linestyle='--',
         marker='o',
         color='tab:green')

ax3.set_xlabel('Year', fontsize=16)
ax3.set_ylabel('Sports', fontsize=16)
ax3.set_xlim([1895, 2017])
ax3.set_xticks(np.arange(1896, 2017, 4))
ax3.set_yticks(np.arange(0,40,5))
ax3.set_ylim([-2,37])
ax3.annotate("WWI", xy=(1915, 32), xycoords='data', fontsize='14')
ax3.annotate('WWII', xy=(1940, 32), xycoords='data', fontsize='14')
ax3.axvspan(1914, 1918, color='grey', alpha=0.15)
ax3.axvspan(1939, 1945, color='grey', alpha=0.15)

plt.title('Number of Sport at the Summer Olympics', fontsize=24)
plt.savefig('number_sport.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Top 10 Gold Medal
goldMedal = dt2[(dt2.Medal == 'Gold')]
```

```python
totalgoldMedal = goldMedal.region.value_counts().reset_index(name='Medal').head(10)
m = sns.catplot(x="index", y="Medal", data=totalgoldMedal,
                height=10, kind="bar", palette="muted")
m.despine(left=True)
m.set_xlabels("Top 10 countries", fontsize=16)
m.set_ylabels("Number of medal", fontsize=16)
plt.title('Top 10 All-time & All-games Gold medal table', fontsize=24)
plt.savefig('top10_gold.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Asian athlete Participation

fig, ax4 = plt.subplots(1, figsize=(24, 6))
ax4.plot(asia_in[['Name','Year']].
         drop_duplicates(
             subset=['Name','Year']).
         groupby('Year').
         count(),
         linestyle='--',
         marker='o',
         color='yellow')

ax4.set_xlabel('Year', fontsize=16)
ax4.set_ylabel('Number of Athlete', fontsize=16)
ax4.set_xlim([1895, 2017])
ax4.set_xticks(np.arange(1896, 2017, 4))
ax4.annotate("WWI", xy=(1915, 265), xycoords='data', fontsize='14')
ax4.annotate('WWII', xy=(1940, 265), xycoords='data', fontsize='14')
ax4.axvspan(1914, 1918, color='grey', alpha=0.15)
ax4.axvspan(1939, 1945, color='grey', alpha=0.15)

ax4.annotate('1964 Tokyo', xy=(1964, 100), xycoords='data',
            xytext=(1964, 150), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            fontsize=18)
ax4.annotate('1988 Seoul', xy=(1988, 160), xycoords='data',
            xytext=(1988, 250), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            fontsize=15)
ax4.annotate('2008 Beijing', xy=(2008, 320), xycoords='data',
            xytext=(2008, 200), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            fontsize=15)

plt.title('Number of Asian athlete at the Summer Olympics', fontsize=24)
plt.savefig('asian_par.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
goldMedalAsia = dt2[(dt2.continent == 'Asia') & (dt2.Medal == 'Gold')]
```

```python
#Top 10 Countries 
goldMedalAsiaSummer = goldMedalAsia[(goldMedalAsia.Season == 'Summer')]
toptenasia = goldMedalAsiaSummer.region.value_counts().reset_index(name='Medal').head(10)
a = sns.catplot(x="index", y="Medal", data=toptenasia,
                height=10, kind="bar")
a.despine(left=True)
a.set_xlabels("Top 10 countries", fontsize=16)
a.set_ylabels("Number of medal", fontsize=16)
plt.title('Top 10 Asian Countries by Gold Medal', fontsize=24)
plt.savefig('asian_top10gold.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Asian Gold by year
fig, ax5 = plt.subplots(1, figsize=(24, 6))
ax5.plot(goldMedalAsiaSummer[['Medal','Year']].
         groupby('Year').
         count(),
         linestyle='--',
         marker='o',
         color='yellow')

ax5.set_xlabel('Year', fontsize=16)
ax5.set_ylabel('Number of Medal', fontsize=16)
ax5.set_xlim([1924, 2017])
ax5.set_xticks(np.arange(1924, 2017, 4))
ax5.annotate('1928 Amsterdam', xy=(1928, 18), xycoords='data',
            xytext=(1928, 65), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            fontsize=15)
ax5.annotate('2008 Beijing', xy=(2008, 142), xycoords='data',
            xytext=(2008, 30), textcoords='data',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='center', verticalalignment='top',
            fontsize=15)

plt.title('Number of Asian Gold Medal at the Summer Games by year', fontsize=24)
plt.savefig('asian_goldbyyear.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Asian Gold by Sex
asian_gold_sex = goldMedalAsiaSummer.Sex.value_counts().reset_index(name='Medal').head(10)
a = sns.catplot(x="index", y="Medal", data=asian_gold_sex,
                height=6, kind="bar", palette=['darkblue','tab:pink'])
a.despine(left=True)
a.set_xlabels("Sex", fontsize=16)
a.set_ylabels("Number of medal", fontsize=16)
plt.title('Number of Asian Gold Medal by Sex', fontsize=24)
plt.savefig('asiangold_bysex.png', dpi=300, bbox_inches='tight')
plt.show()
```
